```python
import pandas as pd
import requests
url = 'http://www.worldometers.info/world-population/population-by-country/'

header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}

r = requests.get(url, headers=header)
df = pd.read_html(r.text)
df
```

```python
WorldPopulation = df[0]
```

```python
WorldPopulation.head()
```

```python
WorldPopulation.info()
```

```python
WorldPopulation = WorldPopulation[['Country (or dependency)','Population (2020)']]
WorldPopulation = WorldPopulation.rename(columns={WorldPopulation.columns[0]:'Country/Territory',WorldPopulation.columns[1]:'Population',})
WorldPopulation
```

```python
WorldPopulation.info()
```

```python
WorldPopulation.to_csv(path_or_buf='Data/WorldPopulation.csv',index=False)
```
