```python
import tabula

pdf_path = 'c:/NEXT/Covid/DailyReports/20200403-WHO.pdf'#set the file path to import here/Definir o caminho para importar aqui

pages = [3,4,5,6,7]#define which pages to import here/defina quais as páginas a importar aqui.

filename1 = 'Data/pages/20200403-WHO_page3.csv'#change the date and page number here/mudar a data e o número de página aqui.
filename2 = 'Data/pages/20200403-WHO_page4.csv'#change the date and page number here/mudar a data e o número de página aqui.
filename3 = 'Data/pages/20200403-WHO_page5.csv'#change the date and page number here/mudar a data e o número de página aqui.
filename4 = 'Data/pages/20200403-WHO_page6.csv'#change the date and page number here/mudar a data e o número de página aqui.
filename5 = 'Data/pages/20200403-WHO_page7.csv'#change the date and page number here/mudar a data e o número de página aqui.

tabula.convert_into(pdf_path,filename1,pages=pages[0],output_format='csv')
tabula.convert_into(pdf_path,filename2,pages=pages[1],output_format='csv')
tabula.convert_into(pdf_path,filename3,pages=pages[2],output_format='csv')
tabula.convert_into(pdf_path,filename4,pages=pages[3],output_format='csv')
tabula.convert_into(pdf_path,filename5,pages=pages[4],output_format='csv')
```

```python
import pandas as pd

df1 = pd.read_csv(filename1,encoding='unicode_escape')
df2 = pd.read_csv(filename2,encoding='unicode_escape')
df3 = pd.read_csv(filename3,encoding='unicode_escape')
df4 = pd.read_csv(filename4,encoding='unicode_escape')
df5 = pd.read_csv(filename5,encoding='unicode_escape')
```

```python
#df1 = df1.dropna(axis=1, how='all')
df1 = df1.rename(columns={df1.columns[0]:'Country/Territory',df1.columns[1]:'Total Confirmed',df1.columns[2]:'New Cases',df1.columns[3]:'Total Deaths',
                        df1.columns[4]:'New Deaths',df1.columns[5]:'Transmission Classification',
                        df1.columns[6]:'Days since last reported case'})
df1 = df1.dropna()
df1 = df1[~df1['Total Confirmed'].str.contains(pat = 'i[a-z]', regex = True)]
df1['Country/Territory'] = df1['Country/Territory'].astype('str')
df1['Total Confirmed'] = df1['Total Confirmed'].astype('int')
df1['New Cases'] = df1['New Cases'].astype('int')
df1['Total Deaths'] = df1['Total Deaths'].astype('int')
df1['New Deaths'] = df1['New Deaths'].astype('int')
df1['Transmission Classification'] = df1['Transmission Classification'].astype('str')
df1['Days since last reported case'] = df1['Days since last reported case'].astype('int')
```
